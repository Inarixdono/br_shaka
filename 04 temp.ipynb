```python
# P29 - 35

tabla = driver.find_element('xpath','//*[@id="MainContent_gvSchooled"]')
cant = len(tabla.find_elements(By.TAG_NAME,'tr')) - 1
cont = 0 

while cont < cant:
    child = driver.find_element('xpath','//*[@id="MainContent_gvSchooled_ddlChildAttendingSchool_'+ str(cont) +'"]')
    repitiendo = driver.find_element('xpath','//*[@id="MainContent_gvSchooled_ddlChildRepeatingGrade_'+ str(cont) +'"]')
    ausente = driver.find_element('xpath','//*[@id="MainContent_gvSchooled_ddlChildStopAttendingSchool_'+ str(cont) +'"]')
    if child.get_attribute('value') != '98':
        Select(child).select_by_index(1)
        Select(repitiendo).select_by_index(2)
        Select(ausente).select_by_index(2)
    cont += 1
```
